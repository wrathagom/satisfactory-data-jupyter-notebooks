```python
import requests as r
import json
from datetime import datetime as dt
import os
import time

from functions import custom
```

```python
startTime = round(time.time() * 1000)
startTime
```

```python
# Borrowed from here: https://stackoverflow.com/questions/5891555/display-the-date-like-may-5th-using-pythons-strftime

def suffix(d):
    return 'th' if 11<=d<=13 else {1:'st',2:'nd',3:'rd'}.get(d%10, 'th')

def custom_strftime(format, t):
    return t.strftime(format).replace('{S}', str(t.day) + suffix(t.day))

endpoints = custom.list_endpoints()
```

```python
for endpoint in endpoints:
    fileName = '../data/raw/{}/snapshot-{}.json'.format(endpoint, custom.custom_strftime('%m%d%Y%H%M', dt.now()))
    print('Saving a Snapshot to "{}"'.format(fileName))
    
    data = r.get('http://ptero.lan:9090/{}'.format(endpoint)).json()
    
    os.makedirs(os.path.dirname(fileName), exist_ok=True)

    with open(fileName, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

    # Recommendation is to not hit the endpoints more than once per second
    time.sleep(1)
```

```python
endTime = round(time.time() * 1000)

totalTime = (endTime - startTime) / 1000
print('Total Execution Time was {} seconds'.format(totalTime))
```
