```python
import json
import pandas as pd
from datetime import datetime as dt

from functions import custom
```

```python
fileName = '../data/raw/getStorageInv/snapshot-1207202505.json'
print('Loading inventory data from "{}"'.format(fileName))
```

```python
f = open(fileName)
data = json.load(f)

f.close()
```

```python
items = {}

for container in data:
    for item in container['Inventory']:
        try:
            items[item['Name']] += item['Amount']
        except KeyError:
            items[item['Name']] = item['Amount']
```

```python
items
```

```python
import matplotlib.pyplot as plt
```

```python
categories = list(items.keys())
values = list(items.values())
```

```python
plt.bar(range(len(items)), values)
plt.xticks(range(len(items)), categories)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart from Dictionary Data')
plt.xticks(rotation=90)
```
